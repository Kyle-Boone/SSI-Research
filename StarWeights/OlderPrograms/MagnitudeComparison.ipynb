```python
import numpy as np
import fitsio
import matplotlib.pyplot as plt
```

```python
starFile = '/afs/hep.wisc.edu/bechtol-group/MegansThings/balrog_detection_catalog_sof_run2_stars_v1.4_avg_added_match_flags.fits'
galaFile = '/hdfs/bechtol/balrog/y3/balrog_detection_catalog_sof_y3-merged_v1.2.fits'
```

```python
starData = fitsio.read(starFile, columns = ['true_g_Corr', 'true_gi_Corr'])
iMagStar = starData['true_g_Corr'] - starData['true_gi_Corr']
galaData = fitsio.read(galaFile, columns = ['true_bdf_mag_deredden'])
iMagGala = galaData['true_bdf_mag_deredden'][:,2]
```

```python
plt.figure(dpi = 150)
plt.grid(zorder = 0)
plt.hist(iMagStar, bins = np.linspace(15, 30, 101), edgecolor='black', linewidth=.5, zorder = 3, density = True, alpha = 0.5, label = 'Stars')
plt.hist(iMagGala, bins = np.linspace(15, 30, 101), edgecolor='black', linewidth=.5, zorder = 3, density = True, alpha = 0.5, label = 'Galaxies')
plt.legend()
plt.show()
```

```python
iMagStarCrop = iMagStar[np.where(iMagStar < 35)[0]]
iMagGalaCrop = iMagGala[np.where(iMagGala < 35)[0]]
```

```python
print(np.average(iMagStarCrop))
print(np.average(iMagGalaCrop))
```
