```python
import numpy as np
import fitsio
import healpy as hp
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

The files are at the path /hdfs/bechtol/pferguso/y3a2_surveyconditions/y3a2_survey_conditions_5_tiles_G.fits

```python
condFile = '/hdfs/bechtol/pferguso/y3a2_surveyconditions/y3a2_survey_conditions_5_tiles_R.fits'
matchedFile = '/afs/hep.wisc.edu/home/kkboone/data/perm/MatchesWithinTiles.fits'
conditionName = 'EXPTIME_SUM'
```

```python
condData = fitsio.read(condFile, columns = ['PIXEL', conditionName])
matchedData = fitsio.read(matchedFile, columns = ['meas_EXTENDED_CLASS_SOF', 'true_ra', 'true_dec'])
```

```python
condPixels = condData['PIXEL']
condition4096 = condData[conditionName]
matchedRa = np.array(matchedData['true_ra'])
matchedDec = np.array(matchedData['true_dec'])
matchedClass = matchedData['meas_EXTENDED_CLASS_SOF']
```

```python
conditionExt = np.zeros(12*(4096**2))
conditionExt[condPixels] = condition4096
mask = np.full(12*(4096**2), True, dtype = bool); mask[condPixels] = False
maskedCond = hp.ma(conditionExt)
maskedCond.mask = mask
```

```python
condition = hp.ud_grade(maskedCond, nside_out = 2048)
```

```python
condition.data
```

```python
plt.figure(dpi = 150)
plt.hist(condition.data, bins = np.linspace(70, 700, 21, endpoint = True))
plt.show()
```

```python
starIndices = np.where((matchedClass == 0) | (matchedClass == 1))[0]
starRa = matchedRa[starIndices]
starDec = matchedDec[starIndices]
```

```python
plt.figure(dpi = 150)
plt.hist(condition, bins = 20)
plt.show()
```

```python
matchedPixels = healpy.ang2pix(4096, starRa, starDec, lonlat = True, nest = True)
```

```python
def binPixelsByCondition(pixels, condition, numBins, min, max):
    binBoundaries = np.linspace(min, max, (numBins + 1))
    pixelsByBin = []
    for i in range(len(binBoundaries) - 1):
        validIndices = np.array(np.where((condition >= binBoundaries[i]) & (condition < binBoundaries[i + 1]))[0])
        validPixels = pixels[validIndices]
        pixelsByBin.append(validPixels)
    return np.array(pixelsByBin)
```

```python
binPixels = binPixelsByCondition(condPixels, condition, 12, 1, 1.4)
```

```python
def matchCountsPerBin(binnedPixels, matchPixels):
    countsByBin = []
    for i in range(len(binnedPixels)):
        counts = 0
        for j in range(len(binnedPixels[i])):
            counts += len(np.where(matchPixels == binnedPixels[i][j])[0])
        countsByBin.append(counts)
    return np.array(countsByBin)
```

```python
binDetectionCounts = matchCountsPerBin(binPixels, matchedPixels)
print(binDetectionCounts)
```

```python
binNumPixels = [len(bin) for bin in binPixels]
print(binNumPixels)
```

```python
binPercent = binDetectionCounts/binNumPixels
binPercent = binPercent/(np.sum(binPercent*binDetectionCounts)/np.sum(binDetectionCounts))#/np.average(binPercent)
print(binPercent)
```

```python
plt.figure(dpi = 150)
plt.plot(np.linspace(1, 1.4, 12), binPercent, marker = '.')
plt.axhline(y=1, color='0.4', linestyle='--')
plt.grid()
plt.show()
```
