```python
import numpy as np
import matplotlib.pyplot as plt
import fitsio
```

```python
veriDir = '/hdfs/bechtol/balrog/y3/y3a2_survey_conditions_maps/Kyle_Stuff/MaximumLikelihood/Verification/'
class0File = '0.5_Class_Star_Count_Histograms.fits'
class1File = '1.5_Class_Star_Count_Histograms.fits'
class2File = '2.5_Class_Star_Count_Histograms.fits'
```

```python
orig0 = fitsio.read(veriDir + class0File)['ORIG']
corr0 = fitsio.read(veriDir + class0File)['CORR']

orig1 = fitsio.read(veriDir + class1File)['ORIG']
corr1 = fitsio.read(veriDir + class1File)['CORR']

orig2 = fitsio.read(veriDir + class2File)['ORIG']
corr2 = fitsio.read(veriDir + class2File)['CORR']
```

```python
# By default it will clump everything beyond the last bin into it, so go one further.
plt.figure(dpi = 120)
plt.hist(np.arange(201), bins = np.linspace(0, 201, 202), weights = orig0, color = 'purple', alpha = 0.5, zorder = 3, label = r'Class $\leq 0$', density=True)
plt.hist(np.arange(201), bins = np.linspace(0, 201, 202), weights = orig1, color = 'orange', alpha = 0.5, zorder = 3, label = r'Class $\leq 1$', density=True)
plt.hist(np.arange(201), bins = np.linspace(0, 201, 202), weights = orig2, color = 'green', alpha = 0.5, zorder = 3, label = r'Class $\leq 2$', density=True)
plt.ylabel('Percent of Pixels')
plt.xlabel('Number of Stars')
plt.title('Distribution of Original Star Counts')
plt.grid(zorder = 0)
plt.legend()
plt.xlim([0, 200])
plt.show()
```

```python
# By default it will clump everything beyond the last bin into it, so go one further.
plt.figure(dpi = 120)
plt.hist(np.arange(201), bins = np.linspace(0, 201, 202), weights = corr0, color = 'purple', alpha = 0.5, zorder = 3, label = r'Class $\leq 0$', density=True)
plt.hist(np.arange(201), bins = np.linspace(0, 201, 202), weights = corr1, color = 'orange', alpha = 0.5, zorder = 3, label = r'Class $\leq 1$', density=True)
plt.hist(np.arange(201), bins = np.linspace(0, 201, 202), weights = corr2, color = 'green', alpha = 0.5, zorder = 3, label = r'Class $\leq 2$', density=True)
plt.ylabel('Percent of Pixels')
plt.xlabel('Number of Stars')
plt.title('Distribution of Corrected Star Counts')
plt.grid(zorder = 0)
plt.legend()
plt.xlim([0, 200])
plt.show()
```
