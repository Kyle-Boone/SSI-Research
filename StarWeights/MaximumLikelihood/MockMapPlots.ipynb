```python
import numpy as np
import fitsio
import matplotlib.pyplot as plt
```

```python
file = '/hdfs/bechtol/balrog/y3/y3a2_survey_conditions_maps/Kyle_Stuff/MaximumLikelihood/Verification/Mock_Map_Residuals.fits'
data = fitsio.read(file)

weight1 = data['1']
weight10 = data['10']
weight100 = data['100']
weight1000 = data['1000']
```

```python
plt.figure(dpi = 120)
plt.hist(np.arange(103) - 51.5, bins = np.linspace(-51.5, 51.5, 104), weights = weight1, alpha = 0.5, zorder = 3, label = '1 Realization')
plt.hist(np.arange(103) - 51.5, bins = np.linspace(-51.5, 51.5, 104), weights = weight10, alpha = 0.5, zorder = 3, label = 'Median of 10 Realizations')
plt.hist(np.arange(103) - 51.5, bins = np.linspace(-51.5, 51.5, 104), weights = weight100, alpha = 0.5, zorder = 3, label = 'Median of 100 Realizations')
plt.hist(np.arange(103) - 51.5, bins = np.linspace(-51.5, 51.5, 104), weights = weight1000, alpha = 0.5, zorder = 3, label = 'Median of 1000 Realizations')
plt.grid(zorder = 0)
plt.xlim([-50, 50])
plt.legend(fontsize = 'small')
plt.xlabel('Difference Between Original Counts and Recovered Counts')
plt.ylabel('Number of Pixels')
plt.title('Residuals of Counts on Mock Realizations')
plt.show()
```
