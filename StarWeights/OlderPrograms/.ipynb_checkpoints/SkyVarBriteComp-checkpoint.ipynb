```python
import numpy as np
import fitsio
import healpy as hp
import matplotlib.pyplot as plt
```

```python
bands = ['g', 'r', 'i', 'z']
condFiles = []
for i in bands:
    condFiles.append('/hdfs/bechtol/balrog/y3/y3a2_survey_conditions_maps/skybright/y3a2_' + i + '_o.4096_t.32768_SKYBRITE.WMEAN_EQU.fits.gz')
for i in bands:
    condFiles.append('/hdfs/bechtol/balrog/y3/y3a2_survey_conditions_maps/skyvar/y3a2_' + i + '_o.4096_t.32768_SKYVAR_WMEAN_EQU.fits.gz')
condFiles = np.array(condFiles)
```

```python
condMaps = []
for i in range(len(condFiles)):
    condMaps.append(hp.fitsfunc.read_map(condFiles[i], nest = True))
    condMaps[i][condMaps[i] == -1.6375e+30] = hp.UNSEEN
condMaps = np.array(condMaps)
```

```python
condSigs = []
for i in condMaps:
    condSigs.append(hp.ud_grade(i,2048))
condSigs = np.array(condSigs)
```

```python
for i in condSigs:
    hp.mollview(i, nest=True, badcolor = 'black')#, cmap = 'viridis')
```
